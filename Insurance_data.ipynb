```python
plt.figure(figsize = (12,5))
sns.boxplot(x = "region", y = "charges", hue = "sex", data = data)
plt.show()
```

For SouthWest, NorthWest, NorthEast charges are almost same for this regions. But SouthEast charges are different compared to other regions.

```python
plt.figure(figsize=(12,4))
sns.barplot(x='age',y='charges',data=data)
plt.show()
```

Age is directly proportional to Charges. If Age increases , Charges increases.

```python
plt.figure(figsize=(20,4))
sns.barplot(x='bmi',y='charges',data=data)
plt.show()
```

If BMI increases then charges also increases.

```python
sns.pairplot(data=data)
plt.show()
```

```python
sns.heatmap(data.corr(),annot=True)
plt.show()
```

Age and BMI will most effect on charges

```python
plt.figure(figsize=(8,4))
sns.barplot(x='sex',y='charges',data=data)
plt.show()
```

Column sex is not much effect on charges because the charges are almost equal for both male and female

```python
plt.figure(figsize = (10,6))
sns.scatterplot(x = "age", y = "charges", data = data, hue = "smoker")
plt.show()
```

Charges that are greater than 30000 are smokers and below 15000 are non-smokers.
